```python
import struct
import numpy as np
# import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
```

```python
#read the dataset files
def read_idx(filename):
    with open(filename, 'rb') as f:
        zero, data_type, dims = struct.unpack('>HBB', f.read(4))
        shape = tuple(struct.unpack('>I', f.read(4))[0] for d in range(dims))
        return np.fromstring(f.read(), dtype=np.uint8).reshape(shape)
    
```

```python
#convert the data to binary
def oneHotEncoding(label):
    n = np.max(label)+1
    v = np.eye(n)[label]
    return v.T
```

```python
def imageProcess(data):
    data = data/255
    data = data.reshape(data.shape[0],data.shape[1]*data.shape[2])
    return data.T
```

```python
#load data
X_train = imageProcess(read_idx('train-images-idx3-ubyte')) 
y_train = oneHotEncoding(read_idx('train-labels-idx1-ubyte'))
X_test = imageProcess(read_idx('t10k-images-idx3-ubyte')) 
y_test = read_idx('t10k-labels-idx1-ubyte')
```

```python
def softMax(X):
    e = np.exp(X)
    p = e/np.sum(e, axis=0)
    return p
```

```python
def ReLU(z):
    return np.maximum(0,z)

def dReLU(z):
    return (z > 0) * 1
```

```python
def sigmoid(z):
    return 1./(1.+np.exp(-z))

def dSigmoid(z):
    return sigmoid(z) *(1-sigmoid (z))
```

```python
def tanh(z):
    return np.tanh(z)

def dTanh(z):
    return 1/(np.cosh(z)**2)
```

```python
#for cost function
def compute_loss(y, y_hat, lamda, params):
    m = y.shape[1]
    loss = -(1/m) * np.sum(y*np.log(y_hat)) + lamda/(2*m) * (np.sum(params['W1']**2) + np.sum(params['W2']**2))
    return loss

```

```python
#forward propogation
def forward_prop(X,params,activation):

    fpass = {}
    fpass['Z1'] = np.matmul(params['W1'], X) + params['b1']
    fpass['A1'] = activation(fpass['Z1'])
    fpass['Z2'] = np.matmul(params['W2'],fpass['A1']) + params['b2']
    fpass['A2'] = softMax(fpass['Z2'])
    return fpass


```

```python
#backward propogation
def backward_prop(X, y,fpass, params,dActivation):
    m = X.shape[1]
    g = {}
    g['dZ2'] = fpass['A2'] - y
    g['dW2'] = (1./m) * np.matmul(g['dZ2'], fpass['A1'].T)
    g['db2'] = (1./m) * np.sum(g['dZ2'], axis=1, keepdims=True)
    g['dA1'] = np.matmul(params['W2'].T, g['dZ2'])
    g['dZ1'] = g['dA1'] * dActivation(fpass['Z1'])
    g['dW1'] = (1./m) * np.matmul(g['dZ1'], X.T)
    g['db1'] = (1./m) * np.sum(g['dZ1'])
    return g

```

```python
#updates the next value
def new_values(params,grad,eta,lamda,m):
    new_val = {}
    new_val['W2'] = params['W2'] - eta * grad['dW2'] - (params['W2']*lamda*eta)/m
    new_val['b2'] = params['b2'] - eta * grad['db2']
    new_val['W1'] = params['W1'] - eta * grad['dW1'] - (params['W1']*lamda*eta)/m
    new_val['b1'] = params['b1'] - eta * grad['db1']
    return new_val
```

```python
def classifer(X, params,activation):
    Z1 = np.matmul(params['W1'], X) + params['b1']
    A1 = activation(Z1)
    Z2 = np.matmul(params['W2'],A1) + params['b2']
    A2 = softMax(Z2)
    pred = np.argmax(A2, axis=0)
    return pred
```

```python
from datetime import datetime

m=10000 #batch size
n_x = X_train.shape[0]
n_h = 100
eta = 1
lamda = 2
np.random.seed(7)
epoch = 1000


sigmoidParams = {'W1': np.random.randn(n_h, n_x)* np.sqrt(1. / n_x),
                 'b1': np.zeros((n_h, 1)),
                 'W2': np.random.randn(10, n_h)* np.sqrt(1. / n_h),
                 'b2': np.zeros((10, 1))
                 }

start = datetime.now()
for i in range(epoch):
    #shuffle batch index
    idx = np.random.permutation(X_train.shape[1])[:m]
    X=X_train[:,idx]
    y=y_train[:,idx]
    #forward pass
    fpass = forward_prop(X,sigmoidParams,sigmoid)
    #cost
    cost = compute_loss(y, fpass['A2'], lamda, sigmoidParams)
    #back Prop
    g = backward_prop(X, y, fpass, sigmoidParams,dSigmoid)
    #updating weights
    sigmoidParams=new_values(sigmoidParams,g,eta,lamda,m)
    
#final cost    
difference = datetime.now() - start
print("Final cost:", cost)
print('time to train:', difference)

y_hat = classifer(X_test, sigmoidParams, sigmoid)


print('Accuracy:',sum(y_hat==y_test)*1/len(y_test))
```

```python
reluParams = {'W1': np.random.randn(n_h, n_x)* np.sqrt(2. / n_x),
                 'b1': np.zeros((n_h, 1)),
                 'W2': np.random.randn(10, n_h)* np.sqrt(2. / n_h),
                 'b2': np.zeros((10, 1))
                 }

start = datetime.now()
for i in range(epoch):
    #shuffle batch index
    idx = np.random.permutation(X_train.shape[1])[:m]
    X = X_train[:,idx]
    y = y_train[:,idx]
    #forward pass
    fpass = forward_prop(X,reluParams,ReLU)
    #cost
    cost = compute_loss(y, fpass['A2'], lamda, reluParams)
    #back Prop
    gradient = backward_prop(X, y, fpass, reluParams,dReLU)
    #updating weights
    reluParams = new_values(reluParams,g,eta,lamda,m)

    #calculate final cost
difference = datetime.now() - start
print("Final cost:", cost)
print('time to train:', difference)

y_hat = classifer(X_test, reluParams, ReLU)


print('Accuracy:',sum(y_hat==y_test)*1/len(y_test))
```
